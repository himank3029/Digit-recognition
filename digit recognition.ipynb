```python
import numpy as np
import pandas as pd
from sklearn import tree,naive_bayes,neighbors
from sklearn.metrics import accuracy_score,confusion_matrix
import time
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
train_data=pd.read_csv('mnist_train.txt')
features_train=np.array(train_data.drop(['label'],'columns')) #dropping label(column) in data
labels_train=np.array(train_data['label'])
```

```python
test_data=pd.read_csv('mnist_test.txt')
features_test=np.array(test_data.drop(['label'],'columns'))
labels_test=np.array(test_data['label'])
```

```python
clf=tree.DecisionTreeClassifier()
#clf=naive_bayes.GaussianNB()
#clf=neighbors.KNeighborsClassifier()
```

```python
#training
t1=time.time()
clf.fit(features_train,labels_train)
t2=time.time()
print("training time:",t2-t1)
```

```python
#testing
t3=time.time()
pre=clf.predict(features_test)
t4=time.time()
print("testing time:",t4-t3)
print("predicted:",pre)
print("actual digit",labels_test)
```

```python
acc=accuracy_score(pre,labels_test)
print("accuracy=",acc)
```

```python
x=69
print("predicted digit:",pre[x])
print("actual digit:",labels_test[x])
digit=features_test[x]
digit_pixels=digit.reshape(28,28)
plt.imshow(digit_pixels,cmap='Accent_r')
plt.show()
```

```python
cm=confusion_matrix(labels_test,pre)
print("confusion matrix")
print(cm)
```

```python
axis=plt.subplot()
sb.heatmap(cm,ax=axis,annot=True)
axis.set_xlabel("predicted digits")
axis.set_ylabel("actual digits")
axis.set_title("decision tree")
sb.heatmap(cm,cmap='cubehelix')
plt.show()
```

```python
clf=naive_bayes.GaussianNB()
#training
t1=time.time()
clf.fit(features_train,labels_train)
t2=time.time()
print("training time:",t2-t1)
```

```python
#testing
t3=time.time()
pre=clf.predict(features_test)
t4=time.time()
print("testing time:",t4-t3)
print("predicted:",pre)
acc=accuracy_score(pre,labels_test)
print("accuracy=",acc)
```
